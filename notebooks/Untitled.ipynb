```python
import numpy as np
# importing matplotlib module 
from matplotlib import pyplot as plt 

%matplotlib notebook
```

```python
sigmoid = lambda x: 1/(1+np.exp(-x))

```

```python
labels = np.array([1,0]).reshape((2,1))
x = np.arange(0,1,0.01).reshape((1,100))
y = labels*np.repeat(np.log(x), repeats=2, axis=0) + (1-labels)*np.repeat(np.log(1-x), repeats=2, axis=0)
y = -np.sum(y, axis=0, keepdims=True)
```

```python
plt.plot(x[0,1:],y[0,1:])
plt.show()
```

```python
x = np.linspace(0.01, 0.99, 50)
y = np.linspace(0.01, 0.99, 50)

X, Y = np.meshgrid(x, y)
```

```python
labels = [1,0]
Z = labels[0]*np.log(X) + (1-labels[0])*np.log(1-X) + labels[1]*np.log(Y) + (1-labels[1])*np.log(1-Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, -Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('surface');
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```
